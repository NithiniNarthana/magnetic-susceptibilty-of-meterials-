```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#for NiCl2 
# Upload the Excel file
file_path = r"C:\Users\narth\Downloads\NiCl2.xlsx"  # Raw string prevents escape sequences
data = pd.read_excel(file_path)

# Extract columns
current = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2]
mean = data['mean']
std = data['std']

current = np.array(current)  # Convert current to a NumPy array
B = 0.7668 * current + 0.1542
square_B = B**2

# Plot graph with error bars
plt.errorbar(square_B, mean * 10**(-3), yerr=std * 10**(-3), fmt='o', label='Data', capsize=5)
plt.xlabel('square of B_field(T)')
plt.ylabel('mass (Kg)')
plt.title('B Field vs c with Error Bars')
plt.grid(True)

# Perform linear regression
mean_scaled = mean * 10**(-3)
slope, intercept = np.polyfit(square_B, mean_scaled, 1)

# Calculate the error in the slope (gradient)
N = len(square_B)
y_fit = slope * square_B + intercept
residuals = mean_scaled - y_fit
residual_std = np.sqrt(np.sum(residuals**2) / (N - 2))

delta_x = square_B - np.mean(square_B)
slope_error = residual_std / np.sqrt(np.sum(delta_x**2))

# Display the slope and its error
print(f"Gradient (slope): {slope:.10f}")
print(f"Error in gradient: {slope_error:.10f}")

# Plot the best-fit line
plt.plot(square_B, y_fit, label=f'Fit: y = {slope:.2e}x + {intercept:.2e}')
plt.legend()
plt.show()
# Calculate the Pearson correlation coefficient (r-value)
r_value = np.corrcoef(square_B, mean_scaled)[0, 1]

# Display the r-value
print(f"R-value (correlation coefficient): {r_value:.4f}")
```
